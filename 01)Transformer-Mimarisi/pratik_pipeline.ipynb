```python
from transformers import pipeline
```

```python
# Duygu Analizi
classifier = pipeline("sentiment-analysis")
```

```python
text = "Bugün hava bana güzel gibi geliyor ama kardeşim böyle düşünmüyor"
```

```python
text2 = "Seni sevmiyor değilim"
```

```python
classifier(text)
```

```python
classifier(text2)
```

```python
import pandas as pd
```

```python
output = classifier(text)
pd.DataFrame(output)
```

```python
# Zero-Shot Sınıflandırma
classifier = pipeline("zero-shot-classification")
```

```python
text = "Sinem Seni Çok Seviyorum"
```

```python
label = ["Aşk", "Deniz", "Gökyüzü"]
```

```python
outputs = classifier(text,label)
pd.DataFrame(outputs)
```

```python
# Metin Üretme
generator = pipeline("text-generation")
```

```python
prompt="Sinem seni çok seviyorum bu mektubu sana yazıyorum"
prompt2 = "Sinem, I love you very much, I am writing this letter to you."
```

```python
outputs = generator(prompt)
output = generator(prompt2,max_length=50)
```

```python
print(outputs)
print(output)
```

```python
generator = pipeline("text-generation",model="distilgpt2")
```

```python
outputs = generator(prompt,max_length=50)
print(outputs[0]["generated_text"])
```

```python
# name entity recognition =
ner = pipeline("ner",grouped_entities=True)
```

```python
text = "My name is Semih from Turkey. HuggingFace is a nice platform"
```

```python
outputs = ner(text)
pd.DataFrame(outputs)
```

```python
# soru cevap analizi
reader = pipeline("question-answering")
```

```python
text = "My name is Semih and I love Italy"
question = "Where do ı like"
```

```python
outputs = reader(question = question , context=text)
pd.DataFrame([outputs])
```

```python
## Metin Özetleme
summarizer = pipeline("summarization")
```

```python
text= "Aşk, insanlık tarihindeki en güçlü ve derin duygulardan biridir. Bu duygu, insanların hayatlarına anlam katan, onları birbirine bağlayan ve yaşamlarını zenginleştiren bir güçtür. Aşk, sadece romantik ilişkilerde değil, aynı zamanda aile bağları, arkadaşlık ve insanlar arasındaki diğer birçok ilişki türünde de kendini gösterir. Bu yazıda, aşkın farklı yönlerini ve etkilerini keşfetmeye çalışacağız. Aşk, genellikle bir romantik ilişkinin temelini oluşturur. İki insan arasındaki romantik aşk, birbirine duyulan derin hisler, çekim ve bağlılık ile karakterizedir. Bu tür bir aşk, insanların birbirine olan güvenini, anlayışını ve fedakarlığını içerir. İki kişi arasındaki bu özel bağ, zorlukları aşmalarını, birlikte büyümelerini ve birbirlerine destek olmalarını sağlar. Aşkın bir diğer yönü, aile bağlarındaki güçlü duygusal bağlantılardır. Aile içindeki sevgi, birçok kişi için yaşamın temel dayanağıdır. Anne-baba sevgisi, kardeşlik bağları ve aile üyeleri arasındaki derin sevgi, insanları bir araya getirir ve onları bir aile olarak birleştirir. Ailedeki bu sevgi ve bağlılık, bireylerin güçlü bir destek sistemine sahip olmalarını sağlar."
```

```python
outputs = summarizer(text, max_length=100,clean_up_tokenization_spaces = True)
```

```python
print(outputs[0]["summary_text"])
```
